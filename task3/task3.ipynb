# PySpark

```python
import pyspark.sql.functions as F
from pyspark import SparkContext
from pyspark import SparkConf
from pyspark import SQLContext

import pandas as pd
```

```python
sc = SparkContext.getOrCreate(SparkConf().setMaster("local[*]"))
sql_sc = SQLContext(sc)
```

```python
df = sql_sc.createDataFrame(pd.read_csv('election.csv'))
```

```python
df.columns
```

```python
new_columns = ['uik', 'tik', 'region', 'n_voters', 'full_papers', 'early_papers', 
               'indoor_papers', 'outdoor_papers', 'cancelled_papers', 'portable_box_papers', 
               'stat_box_papers', 'invalid_papers', 'valid_papers', 'lost_papers', 'uncouted_papers',
              'baburin', 'grudinin', 'zhirinovskyi', 'putin', 'sobchak', 'suraikin', 'titov', 'yavlinskyi']
```

```python
df = df.toDF(*new_columns)
```

```python
df.columns
```

```python
df.toPandas().head()
```

### 1. Явка (%) по всем регионам

```python
(df.groupBy('region').agg(
    F.sum('n_voters').alias('sum_n_voters'),
    F.sum('early_papers').alias('sum_early_papers'),
    F.sum('indoor_papers').alias('sum_indoor_papers'),
    F.sum('outdoor_papers').alias('sum_outdoor_papers')
).withColumn('total', 
             (F.col('sum_early_papers') 
              + F.col('sum_indoor_papers') 
              + F.col('sum_outdoor_papers')) / F.col('sum_n_voters')) 
.sort('total', ascending=False)).show(100)
```

### 2. Выбрать кандидата и найти избиратльный участок, на котором он получил наибольший результат (учитывать участки на которых проголосовало больше 300 человек)

```python
candidate = "grudinin"

(df.withColumn('total', F.col('early_papers')  + F.col('indoor_papers') + F.col('outdoor_papers'))
                  .withColumn('percent', F.col(candidate) / F.col('total'))
                  .filter(F.col('total') > 300)
                  .select('region', 'tik', 'uik', candidate, 'total', 'percent')
                  .sort('percent', ascending=False).show(1))
```

### 3. Найти регион, где разница между ТИК с наибольшей явкой и наименьшей максимальна 

```python
(df.withColumn('total', F.col('early_papers') + F.col('indoor_papers') + F.col('outdoor_papers'))
    .groupBy('region', 'tik')
    .agg(F.sum('total').alias('tik_total'), F.sum('n_voters').alias('tik_n_voters'))
    .withColumn('tik_percent', F.col('tik_total') / F.col('tik_n_voters'))
    .groupBy('region')
    .agg(F.max('tik_percent').alias('max'), F.min('tik_percent').alias('min'))
    .withColumn('difference', F.col('max') - F.col('min'))
    .sort('difference', ascending=False)
).show(1)
```

### 4. Посчитать дисперсию по явке для каждого региона (учитывать УИК) 

```python
(df.withColumn('total', 
               (F.col('early_papers') + F.col('indoor_papers') + F.col('outdoor_papers')) / F.col('n_voters'))
    .groupBy('region')
    .agg(F.stddev('total').alias('std'))
    .sort('std')
).show(100)
```

###  5. Для каждого кандидата посчитать таблицу: результат (%, округленный до целого) - количество УИК, на которых кандидат получил данный результат

```python
candidates = ['baburin', 'grudinin', 'zhirinovskyi', 'putin', 'sobchak', 'suraikin', 'titov', 'yavlinskyi']
```

```python
df = df.withColumn('n_votes', df['portable_box_papers'] +  df['stat_box_papers'])
```

```python
for candidate in candidates:
    print(candidate)
    (df.withColumn('result', F.round(100 * df[candidate] / df['n_votes']).cast('int'))
        .na.fill(0) # because there is one record with zero votes
        .groupBy('result')
        .count()
        .sort('result')
        .show(10))
```
